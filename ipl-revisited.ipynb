```python
def extract_info(file_name): ##Removed DLS Matches and Super Over Matches
    file=open(file_name)
    data=json.load(file)
    info={}
    if "winner" not in data["info"]["outcome"]:
        return
    if "method" in data["info"]["outcome"] and data["info"]["outcome"]["method"]=="D/L":
        return
    info["Match_id"]=str(file_name).split("/")[-1][:-5]
    city_name=""
    if "city" in data["info"]:
        city_name=","+data["info"]["city"]
    info["Venue"]=data["info"]["venue"]+city_name
    info["Team1"]=data["info"]["teams"][0]
    info["Team2"]=data["info"]["teams"][1]
    info["Winner"]=data["info"]['outcome']["winner"]
    file.close()
    return info
```

```python
def extract_match_data(file_name):
    file=open(file_name)
    data=json.load(file)
    match_data=pd.read_csv("/kaggle/input/ipl-data-processed/Matches.csv")
    match_data["Match_id"]=match_data["Match_id"].astype(str)
    match_id=file_name.split("/")[-1][:-5]
    if match_id in ["1136564","335994","392198"]:
        file.close()
        return
    final_df=pd.DataFrame(columns=["Match_id","batter","non_striker","batting_team","bowling_team","over","ball","target","wickets"])
    for i in range(len(data["innings"][1]["overs"])):
        over=i
        counter=[1,2,3,4,5,6]
        idx=0
        df=pd.json_normalize(data["innings"][1]["overs"][over]["deliveries"])
        df["Match_id"]=match_id
        df["over"]=over
        df["ball"]=-1
        df["batting_team"]=data["innings"][1]["team"]
        df["bowling_team"]=data["innings"][0]["team"]
        df["target"]=data["innings"][1]["target"]["runs"]
        if "wickets" not in df.columns:
            df["wickets"]=0
        if "extras.noballs" not in df.columns:
            df["extras.noballs"]=0
        if "extras.wides" not in df.columns:
            df["extras.wides"]=0

        df["wickets"]=df["wickets"].fillna(0)
        df["wickets"]=df["wickets"].apply(lambda x:x if x==0 else 1)

        df["extras.noballs"]=df["extras.noballs"].fillna(0)
        df["extras.noballs"]=df["extras.noballs"].apply(lambda x:x if x==0 else 1)

        df["extras.wides"]=df["extras.wides"].fillna(0)
        df["extras.wides"]=df["extras.wides"].apply(lambda x:x if x==0 else 1)

        df=df.drop(columns=["runs.batter","runs.extras"])
        for i in df.index:
            if df["extras.noballs"][i] or df["extras.wides"][i]:
                df.loc[i,"ball"]=counter[idx]
            else:
                df.loc[i,"ball"]=counter[idx]
                idx+=1
        
        idx=0
        df=df[["Match_id","batter","non_striker","batting_team","bowling_team","over","ball","runs.total","target","wickets"]]
        df["runs_scored"]=df["runs.total"]
        df=df.drop(columns=["runs.total"])
        final_df=pd.concat([final_df,df],ignore_index=True)
    
    final_df["balls_bowled"]=final_df["over"]*6+final_df["ball"]
    final_df["balls_left"]=120-final_df["balls_bowled"]
    
    final_df["wickets"]=final_df["wickets"].cumsum()
    final_df["wickets_left"]=10-final_df["wickets"]
    
    final_df["runs_scored"]=final_df["runs_scored"].astype(int)
    final_df["total_runs"]=final_df["runs_scored"].cumsum()
    final_df["runs_left"]=final_df["target"]-final_df["total_runs"]
    final_df.loc[final_df["runs_left"]<0,"runs_left"]=0
    
    final_df["crr"]=final_df["total_runs"]*6/final_df["balls_bowled"]
    final_df["rrr"]=final_df["runs_left"]*6/(final_df["balls_left"]+10**(-2))
    
    final_df=final_df.drop(columns=["over","ball","balls_bowled"])
    final_df=final_df.drop(columns=["runs_scored","total_runs"])
    final_df=final_df.drop(columns=["wickets"])
    
    final_df=pd.merge(final_df,match_data[["Match_id","Venue","Winner"]],on="Match_id")
    final_df.loc[final_df["batting_team"]==final_df["Winner"],"Winner"]=1
    final_df.loc[final_df["bowling_team"]==final_df["Winner"],"Winner"]=0
    file.close()
    
    return final_df
```

```python
import requests
from bs4 import BeautifulSoup
import dateparser

def get_career_averages(idt=277916,file_name=None, match_format=11, data_type='allround') :

        """Get Player career averages

        Arguements:
            file_name {string}: File name to save data
            match_format {int}: Match format (default is 11) (1-Test), (2-Odi) (3-T20I), (11-All International), (20-Youth Tests), (21-Youth ODI)
            data_type {string}: Data type (default is allround) (allround, batting, bowling, fielding)
        
        Return:
            Data in csv file
        """
        player_id=idt
        url = "https://www.espncricinfo.com/player/player-name-{0}".format(str(player_id))
        headers = {'user-agent': 'Mozilla/5.0'}
        match_format = match_format
        data_type = data_type
        #self.file_name = file_name

        #if self.file_name is None:
         #   self.file_name = f"{self.player_id}_{self.match_format}_{self.data_type}_career_averages.csv"

        url=f"https://stats.espncricinfo.com/ci/engine/player/{player_id}.html?class={match_format};template=results;type={data_type}"
        html_doc = requests.get(url, headers=headers)
        soup = BeautifulSoup(html_doc.text, 'html.parser')
        tables = soup.find_all("table")[2]
        table_rows = tables.find_all("tr")
        scores =[]
        for tr in table_rows:
            scores.append(tr.text)
            
        return scores,tables,table_rows
        
    
```

```python
def get_strike_rate(player_id):
    headers = {'user-agent': 'Mozilla/5.0'}
    url=f"https://stats.espncricinfo.com/ci/engine/player/{player_id}.html"
    webpage=requests.get(url,headers=headers).text 
    soup=BeautifulSoup(webpage,"lxml")
    values,results=[],[]
    for i in soup.find_all("p",class_="ds-text-title-subtle-s ds-font-medium ds-py-2 ds-ml-4 ds-text-typo ds-capitalize"):
        s=i.get_text("|",strip=True)
        values.append(s)

    for i in soup.find_all("tr"):
        s=i.get_text("|",strip=True)
        l=s.split("|")
        if l[0]=="T20s":
            if len(s)>=9:
                results.append(l[8])
            
    s=results[values.index("Batting & Fielding")]
    if s=="-":
        return "0"
    else:
        return s
```
